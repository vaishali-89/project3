```python
import pandas as pd

```

Read the Excel file: Use pandas.read_excel() to load the data into a DataFrame.

```python
# Load the Excel file
file_path = 'learntube_ds_project.xlsx'  # Replace with your actual file path
sheet_name = 'Project 3'  # Replace with your actual sheet name if needed
data = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
```

```python
data.head()
```


The dataset contains three tables in a single sheet. Below is the step-by-step approach to separate them:

Table Identification:

Table 1: Starts at row 2 and ends at row 11, columns 0 to 7.
Table 2: Starts at row 2 and ends at row 11, columns 9 to 15.
Table 3: Starts at row 2 and ends at row 11, columns 24 to 30.
Extracting Tables: We will use the iloc function to extract each table.

```python
# Table 1
table1= data.iloc[2:38, 0:8]
table1.columns = ['Machine_ID', 'Date', 'Spindle_Speed_RPM', 'Vibration_Level (mm/s)', 
                  'Temperature (°C)', 'Operation_Hours', 'Machine_Status', 'Predicted_Failure']

# Table 2
table2 = data.iloc[2:53, 9:16]
table2.columns = ['Machine_ID', 'Failure_Date', 'Failure_Type', 'Downtime_Hours', 
                  'Repair_Time (hours)', 'Issue_Resolved_By', 'Comments']
# Table 3 - Maintenance Schedule
table3 = data.iloc[1:48, 17:23]
table3.columns = ['Machine_ID', 'Last_Maintenance_Date', 'Next_Schedule_Maintenance_Date', 'Maintenance_Type', 'Technician_Name', 'Maintenance_Comments']

# Table 4 - Machine Usage
table4 = data.iloc[1:48, 24:31]
table4.columns = ['Machine_ID', 'Usage_Date', 'Parts_Produced', 'Production_Quality', 'Rejection_Rate', 'Idle_Time', 'Operation_name']


# Display the extracted tables
table1.head(), table2.head(), table3.head(), table4.head()
```

```python
# Step 2: Merge tables on 'Machine_ID'
merged_data = pd.merge(table1, table2, on='Machine_ID', how='left')
merged_data = pd.merge(merged_data, table3, on='Machine_ID', how='left')
merged_data = pd.merge(merged_data, table4, on='Machine_ID', how='left')

# Display the final merged DataFrame
merged_data.head()
```
